```python
alltrips["from_station_id"] = alltrips["from_station_id"].astype(object)
alltrips["to_station_id"] = alltrips["to_station_id"].astype(object)
alltrips["trip_id"] = alltrips["trip_id"].astype(object)
alltrips["bikeid"] = alltrips["bikeid"].astype(object)
alltrips["birthyear"] = alltrips["birthyear"].astype(int)
```

```python
alltrips["tripduration"] =  pd.to_numeric(alltrips["tripduration"], errors='coerce')
```

```python
alltrips['age'] = 2022 -  alltrips['birthyear']
```

```python
alltrips['day'] = alltrips.start_time.apply(lambda x: x.day_name())  
alltrips['month'] = alltrips.start_time.apply(lambda x: x.month)
alltrips['year'] = alltrips.start_time.apply(lambda x: x.year)
alltrips['hour'] = alltrips.start_time.apply(lambda x: x.hour)
```

```python
import datetime as datetime
from datetime import timedelta
```

```python
alltrips['total_ride_time'] = (alltrips['end_time'] - alltrips['start_time'])
alltrips['total_ride_time'] = (alltrips['total_ride_time'])/timedelta(minutes=1)
alltrips['total_ride_time'] = alltrips['total_ride_time'].round(decimals = 1)
```

```python
alltrips['usertype'].replace({'Subscriber':'casual', 'Customer':'member'}, inplace=True)
```
